#### Data Cleaning 

* Key steps in data cleaning
* Data Inspection
* Handling missing data
  1. Delection of rows or colomns
  2. imputation method: This is the process of filling the missing value by the mode, median and the mean
     * Method of impution
     - .fillna: To fill a missing values (blank)
     - .dropna: This is use to drop any missing values(blank)
     - K-NN method: Automatically filling of missing values(Machine learning)
* Drop: Deleting the entire column from the data   

#### Handling the missing values

```python
df.isnull().sum()
```

```python
df["color"].isnull().sum()
```

```python
df["color"].value_counts()
```

```python
df["color"] = df["color"].fillna("Black")
```

```python
df["mileage"].isnull().sum()
```

```python
average_milleage = round(df["mileage"].mean(),1)

average_milleage
```

```python
df["mileage"] = df["mileage"].fillna(average_milleage)
```

```python
df["engine_size"].isnull().sum()
```

```python
average_engine_size = round(df["engine_size"].mean(),1)
```

```python
df["engine_size"] = df["engine_size"].fillna(average_engine_size)
```

```python
df["selling_cond"].isnull().sum()
```

```python
df.dropna(subset = "selling_cond", inplace = True )
```

```python
df.dropna(subset = "bought_cond", inplace = True)
```

```python
df.isnull().sum()
```

```python
df.drop(columns = ["trim", "drive_train", "reg_city", "seat", "num_cylinder", "horse_power", "body_build"], inplace = True)                           
```

```python
df.dropna(subset = "fuel_type", inplace = True)
df.dropna(subset = "transmission", inplace = True)
```

```python
# Number of cars sold by the region

Top_6_region_by_sales = df.groupby("region")["car_id"].count().reset_index().sort_values(by = "car_id", ascending = False)[:6]
```

```python
Top_6_region_by_sales
```

* Data Visualization with seaborn and matplotlib

```python
plt.figure(figsize = (10,5))
sns.barplot(data = Top_6_region_by_sales, x = "region", y = "car_id", palette = "magma")
plt.title("Top 6 Sales by Region Analysis")
plt.xlabel("Region")
plt.ylabel("Number of Cars")
plt.xticks(rotation = 20)
plt.show()
plt.savefig("Top 6 Sales by Region.png")
```

```python
 Top_5_Cars_Make = df.groupby("make")["car_id"].count().reset_index().sort_values(by = "car_id", ascending = False)[:5]
```

```python
plt.figure(figsize = (10,5))
sns.barplot(data = Top_5_Cars_Make, x = "make", y = "car_id", palette = "magma")
plt.title("Top 5 Sales by Car Make Analysis")
plt.xlabel("Make")
plt.ylabel("Number of Cars")
plt.xticks(rotation = 20)
plt.show()
```

```python
Top_4_Sales_by_fueltype = df.groupby("fuel_type")["car_id"].count().reset_index().sort_values(by = "car_id", ascending = False)[:4]
```

```python
plt.figure(figsize = (10,5))
sns.barplot(data = Top_4_Sales_by_fueltype, x = "fuel_type", y = "car_id", palette = "magma")
plt.title("Top 4 Sales by Fuel Type Analysis")
plt.xlabel("Fuel_Type")
plt.ylabel("Number of Cars")
plt.xticks(rotation = 20)
plt.show()
```

```python
sales_trend = df.groupby("year_of_man")["amount"].sum().reset_index().sort_values(by = "amount", ascending = False)
```

```python
plt.figure(figsize = (10,5))
sns.lineplot(x="year_of_man", y="amount", data= sales_trend, color = "purple")
plt.title("Sales Trend")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```
